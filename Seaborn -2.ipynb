## Categorical plot

### Countplot

```python
sns.countplot(x = "day", data = tips)
plt.show()
```

```python
fig, ax = plt.subplots()

ax = sns.countplot(x = "day", data = tips)

for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x() + 0.3, p.get_height()+ 0.5))
    
plt.show()
```

```python
print(ax.patches)
```

```python
for p in ax.patches:
    print(p)
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

ax = sns.countplot(x = "day", data = tips, hue = "sex")

for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x() + 0.15, p.get_height() + 1))
    
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

ax = sns.countplot(x = "day", data = tips, hue = "sex")

for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x() + 0.15, p.get_height() + 1))
    ax.text(p.get_x() + 0.07, p.get_height()-4, str(round(p.get_height()/tips.day.count()*100, 2)))
    
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

ax = sns.countplot(x = "day", data = tips, hue = "sex")

for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x() + 0.15, p.get_height() + 1))
    ax.text(p.get_x() + 0.07, p.get_height()-4, "%"+ str(round(p.get_height()/tips.day.count()*100, 2)))
    
plt.show()
```

## Pair Grid

```python
g = sns.PairGrid(tips, hue = "sex", palette = "rocket")

g = g.map_upper(sns.scatterplot, s = 200)
g = g.map_diag(sns.histplot)
g = g.map_lower(sns.kdeplot, shade = True, color = "g")

g = g.add_legend();
```
